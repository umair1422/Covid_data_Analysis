```python
import numpy as np
import pandas as pd
import sqlite3
import csv
import matplotlib.pyplot as plt
```

```python
covidData = pd.read_csv("C:/Users/umair/Downloads/WHO-COVID-19-global-data.csv")
```

```python
database_name = 'CovidDB'  # Replace with your desired database name
conn = sqlite3.connect(database_name)
table_name = 'Covid19tbl'  # Replace with your desired table name
covidDatasql=covidData.to_sql(table_name, conn, if_exists='replace', index=False)
```

```python
cursor = conn.cursor()

# Execute an SQL query to retrieve data
query = f"SELECT country,SUM(New_cases), SUM(New_deaths) AS average_death FROM {table_name} GROUP BY country;"
cursor.execute(query)

# Fetch all rows from the result
rows = cursor.fetchall()

# Print the data
for row in rows:
    print(row)

    
```
